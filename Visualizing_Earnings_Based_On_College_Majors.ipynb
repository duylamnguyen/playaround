# Visualize Earnings Based on College Majors

In this project, we will answer below questions using visualizations:

- Do students in more popular majors make more money?
    * Using scatter plots
- How many majors are predominantly male? Predominantly female?
    * Using histogram
- Which category of majors have the most students?
    * Using bar plots
  
We will use this cleaned [data set](https://github.com/fivethirtyeight/data/tree/master/college-majors) from FiveThirtyEight. The original data was released by [American Community Survey](https://www.census.gov/programs-surveys/acs/).

Let's set up the environment by importing necessary libraries and read the data set.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

recent_grads = pd.read_csv('D:/Data Science/Dataset/recent-grads.csv')
print(recent_grads.iloc[0])
print()
print(recent_grads.head())
print()
print(recent_grads.tail())
print()
print(recent_grads.describe())
```

```python
raw_data_count = recent_grads.shape[0]
print(raw_data_count)
```

```python
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape[0]
print(cleaned_data_count)
```

Only one row has missing values. We need to remove missing values from data set or matplotlib will throw error because the lengh of columns of values we pass in do not match each others.
